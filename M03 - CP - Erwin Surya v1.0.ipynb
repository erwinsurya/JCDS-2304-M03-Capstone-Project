## 6. Feature Selection

```python
# mengecek normality dari kolom-kolom numerik

normality = X_train.select_dtypes(include = 'number').apply(lambda col: normaltest(col.dropna())[1]).reset_index()
normality.columns = ['Column', 'p-value']
normality
```

```python
# menampilkan matriks korelasi antar fitur numerik

plt.figure(figsize = (8, 6))
sns.heatmap(X_train.select_dtypes(include = 'number').corr(method = 'spearman'), annot = True)
plt.show()
```

```python
# mengecek multicollinearity

dfcorr = X_train.select_dtypes(include = 'number')
corrMatrix = dfcorr.corr(method = 'spearman')
already_in = set()
result = []
for col in corrMatrix:
    large_corr = corrMatrix[col][corrMatrix[col] >= 0.6 ].index.tolist()
    if large_corr and col not in already_in:
        already_in.update(set(large_corr))
        large_corr.append(col)
        result.append(large_corr)
result
```

```python
# feature importance

X_train_feat = X_train.copy()
categorical = X_train_feat.select_dtypes(include = 'object').columns
X_train_feat[categorical] = X_train_feat[categorical].fillna(value = 'missing')

le = LabelEncoder()
X_train_feat[categorical] = X_train_feat[categorical].astype('string').apply(le.fit_transform)

xgb = XGBClassifier()
xgb.fit(X_train_feat, y_train)
xgb_coef1 = pd.Series(xgb.feature_importances_, X_train_feat.columns).sort_values(ascending = False)

plt.figure(figsize = (8, 6))
xgb_coef1.iloc[:50].plot(kind = 'bar', title = 'Feature Importance')
plt.show()
```

## 7. Modelling

```python
# menyiapkan pipeline untuk preprocessing numerik dan kategorik

preprocessor = ColumnTransformer(
    transformers=[
        ('Numerical 1', Pipeline([('scaler', RobustScaler())]), ['Duration', 'Sales', 'Commission']),
        ('Numerical 2', Pipeline([('scaler', StandardScaler())]), ['Age']),
        ('Categorical 1', Pipeline([('encoder', OneHotEncoder(drop = 'first'))]), ['Type', 'Channel']),
        ('Categorical 2', Pipeline([('encoder', BinaryEncoder())]), ['Agency', 'Product', 'Subregion'])
    ]
)

pipeline = ImbPipeline(steps=[
    ('preprocessor', preprocessor),
    ('feature_selection', SelectKBest(score_func = f_classif)),
    ('resampling', RandomUnderSampler(random_state = 2024)),
    ('classifier', RandomForestClassifier(random_state = 2024))
])

param_grid = [
    {
        'feature_selection__k': [5, 9],
        'classifier': [LogisticRegression(max_iter = 1000)],
        'classifier__C': [0.01, 0.1, 1, 10, 100],
        'classifier__solver': ['lbfgs', 'liblinear']
    },
    {
        'feature_selection__k': [5, 9],
        'classifier': [DecisionTreeClassifier()],
        'classifier__max_depth': [None, 10, 20],
        'classifier__min_samples_split': [2, 5],
        'classifier__min_samples_leaf': [1, 4]
    },   
    {
        'feature_selection__k': [5, 9],
        'classifier': [RandomForestClassifier()],
        'classifier__n_estimators': [50, 100, 200],
        'classifier__max_depth': [None, 10, 20],
        'classifier__min_samples_split': [2, 5],
        'classifier__min_samples_leaf': [1, 4]
    },
    {
        'feature_selection__k': [5, 9],
        'classifier': [XGBClassifier()],
        'classifier__n_estimators': [50, 100, 200],
        'classifier__max_depth': [None, 10, 20],
        'classifier__learning_rate': [0.01, 0.1],
        'classifier__subsample': [0.6, 0.8]
    },
]

# membuat GridSearchCV dengan pipeline
gscv = GridSearchCV(pipeline, param_grid, cv = 5, n_jobs = -1, scoring = 'f1')

# fit GridSearchCV ke data
gscv.fit(X_train, y_train)
```
